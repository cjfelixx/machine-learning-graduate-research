# Kernel Recursive Least Squares (KRLS) & Kernel Least Mean Squares

Example: Time series prediction

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500

# AWGN
v = np.random.normal(0,1,N) 
# Initial conditoins
d =[0.1, 0.1]

d[0] = d[0] + v[0]
d[1] = d[1] + v[1]

# Discrete time system
for k in range(2,N):
#     print(d[k-1])
    d.append((0.8 - 0.5 * np.exp(-(d[k-1]**2)))*d[k-1] 
             - (0.3 + np.exp(-(d[k-1]**2)))*d[k-2] 
             + 0.1*np.sin(np.pi*d[k-1])
            + v[k])
```

```python
plt.plot(list(range(N)),d)
```

```python
import numpy as np
import matplotlib.pyplot as plt

phase = np.zeros((50,100,100))


for g in range(0,1):
    r1 = np.random.randint(50) * 2
    r2 = np.random.randint(50) * 2
    
    
    noise = np.random.normal(0,1,100)
    
    for y in range(0,100):
        for x in range(0,100):
    
            phase[g,y,x] = r1 * (x ** 4) + r2 * (x ** 3) * (y + 1) + r1 * (x ** 2) * ((y + 1) ** 2) + r2 * ((y + 1) ** 4)

        phase[g,y] = [sum(i) for i in zip(noise, phase[g,y])]
```
